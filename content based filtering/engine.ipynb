```python
import pandas as pd
import numpy as np
```

```python
df_credit = pd.read_csv("E:\GITHUB\Recomendation_engine_tryout/tmdb_5000_credits.csv")
df_movies = pd.read_csv("E:\GITHUB\Recomendation_engine_tryout/tmdb_5000_movies.csv")
```

```python

# joining the data sets on the basis of movie_ID column

new_credit_df = df_credit.rename(index= str,columns={"movie_id":"id"})
df_movies_merge = df_movies.merge(new_credit_df,on='id')
df_movies_merge.head()
```

```python
df_Cleaned = df_movies_merge.drop(columns=['homepage','production_countries', 'status', 'title_x','title_y'])
df_Cleaned.head()
```

# now we are making a recommendation system based on the movie plot summaries give in the overview column. So if we come across a user giving us
# a movie title we recommend titles with similar plots

```python
df_Cleaned.head()[['original_title','overview']]
```

# use of tvidf NLP process converting all the sentences (strings) in the overview columns and convert them into document matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df= 3 , max_features= None,
                      strip_accents= 'unicode',analyzer= 'word',token_pattern= r'\w{1,}',
                      ngram_range= (1,3),
                      stop_words= 'english')

#filing nans with empty strings

df_Cleaned['overview'] = df_Cleaned['overview'].fillna('')
```

```python
tvf_matrix = tfv.fit_transform(df_Cleaned['overview'])
```

```python
tvf_matrix.shape
```

```python
tvf_matrix
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tvf_matrix,tvf_matrix)
```

```python
sig
```

```python
#reverse mapping of indices and movie titles

indices = pd.Series(df_Cleaned.index , index= df_Cleaned['original_title']).drop_duplicates()
```

```python
indices
```

```python
def give_recommendation( title , sig = sig):

     # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return df_Cleaned['original_title'].iloc[movie_indices]
```

```python
give_recommendation("Soldier")
```

```python
give_recommendation("Alien")
```
